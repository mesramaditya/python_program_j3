```python
# dict in data structure
1. dict is collection of item enclosed {} - here item is a key:value pair
2. in dict key are unique
```

```python
# creat dict
info={"name":"shiva","add":"hyd","salary":15000,"age":25}
print(info)
print(type(info))
```

```python
# accesing value from dict using key
info={"name":"shiva","add":"hyd","salary":15000,"age":25}
print(info)
print(info["salary"])
```

```python
#methodor function in dict
print(dir(dict))
```

```python
#keys(),values(),items()
info={"name":"shiva","add":"hyd","salary":15000,"age":25}
print(info)
print(info.keys())
print(info.values())
print(info.items())
```

```python
# update()
info={"name":"shiva","add":"hyd","salary":15000,"age":25}
print(info)
info["add"]="hyderabad"
print(info)
info["domine"]="data enginee"
print(info)
```

```python
#pop()
info={"name":"shiva","add":"hyd","salary":15000,"age":25}
print(info)
info.pop("add")
print(info)
```

```python
# sets in data structures
1. set in collection of unorder unique elementes enclosed by {}
```

```python
#creat set 
a=(70,50,15,30,20,10,10,30,15)
print(a)
print(type(a))
```

```python
#methods in set
print(dir(set))
```

```python
#add() - addelements to set
a={10,20,30,40,60}
print(a)
a.add(100)
print(a)
```

```python
#difference
a={1,2,3,4,5}
b={3,4,5,6,8,}
c=a.difference(b)
print(c)
```

```python
#union)()
a={1,2,3,4,5}
b={3,4,5,6,8,}
c=a.union(b)
print(c)
```

```python
#intersection
a={1,2,3,4,5}
b={3,4,5,6,8,}
c=a.intersection(b)
print(c)
```

```python
#discart()
a={10,20,30,40,50}
print(a)
a.discard(30)
print(a)
```
