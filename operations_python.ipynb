# Operations in python

## 6 types of operations in python
1. arithematic operators
2. assingment operators
3. comparision operators
4. logical operators
5. membership operators
6. identity operators

```python
# arthemetic operators --- +,-,*,/,//,%
a=10
b=20
c=20
d=30
print(a+b)
print(b-c)
print(b*d)
print(d/a)
print(d//a)
print(d%a)
```

```python
#assingment operators --- +=,-=,*=,/=,//=,%=
a=100
a+=50
print(a)
```

```python
b=200
b-=400
print(b)
```

```python
c=50
c*=100
print(c)
```

```python
# comparision operators-- >,<,>=,<=,==,!=
a=10
b=20
c=20
d=30
print(a>b)
print(b==c)
print(b<c)
print(b!=c)
print(b>=a)
print(b<=a)
```

```python
# membership operators - in,not in
#validating if the members are present in tht sequence or group
names=['vikas','shivaji','sidh','chandra']
print('vikas' in names)
print('shivaji' not in names)
print('Vikas' not in names)
print('vikram' in names)
```
