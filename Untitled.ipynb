```python
#write a program to print 1 to 100 numbers using for loop and while loop ?
for num in range(1,101):
    print(num,end=",")
```

```python
sn=1 
en=100 
while (sn<=en):
    print(sn,end=",")
    sn=sn+1
```

```python
#write a program to print even numbers in between 1 to 100 using for loop and while loop?
for num in range (1,101): 
    if (num%2==0):
        print(num,end="<")
```

```python
#write a program to find the sum of all even numbers in between 1 to 100 using loop and while loop 
sum_even=0
for i in range(1,101):
    if (i%2==0):
        sum_even=sum_even+i     
print("sum of all even numbers is ",sum_even)
```

```python
sn=1 
en=100
se=0
while(sn<=en):
    if (sn%2==0):
        se=se+sn
    sn=sn+1
print("sum of all even numbers is",se)        
```

```python
#write a program to find the sum of all odd numbers in between 1 to 100 using loop and while loop 
sum_odd=0
for i in range(1,101):
    if (i%2==1):
        sum_odd=sum_odd+i     
print("sum of all odd numbers is ",sum_odd)
```
